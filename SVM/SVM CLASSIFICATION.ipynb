```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\ayush\Downloads\placement.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x='cgpa',y='placement_exam_marks',data=df,hue='placed')
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df['placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='rbf')
svc.fit(x_train,y_train)
```

```python
svc.score(x_test,y_test)*100
```

```python
svc.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svc)
plt.show()
```
